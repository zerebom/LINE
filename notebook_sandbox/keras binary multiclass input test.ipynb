```python
def weight_method(X, Y, nclasses):
    _Y = np.argmax(Y, axis=4)
    counts = np.array([(_Y == label).sum() for label in range(nclasses)])
    counts = np.cbrt(np.where(counts != 0, counts, 1))
    Y = Y * 1e0 / counts
    return Y
```
